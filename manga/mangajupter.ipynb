```python
import os,json
from PIL import Image

directory = '.'  # 替换为实际的目录路径   
# print(json.dumps(image_paths, indent=4,ensure_ascii=False))
```

```python
def get_image_paths(directory):
    image_paths = []
    for root, dirs, files in os.walk(directory):   
        for file in files:
            # 检查文件扩展名是否为 JPEG 或 PNG
            if file.lower().endswith(('.jpeg', '.jpg', '.png')):    
                image_paths.append(os.path.join(root, file)) 
        #对文件名进行排序    
        image_paths.sort()          
    return image_paths
```

```python
get_image_paths(directory)
```
