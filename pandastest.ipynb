```python
frame = pd.read_excel(Xlsx,'按人员统计代码提交分析')
```

```python
def assessment(attendv=0,repov=0,codcommitv=0):
    assess_total = 0
    repo_total = 0
    #参与项目评分
    attentscore = {1:3,2:5,3:7,4:9,5:11,6:13,7:16,8:19,9:22,10:25,11:28}
    if 0<attendv<12:
        assess_total = attentscore[attendv]
    elif attendv>=12:
        assess_total = 30
        
    #参与代码库评分
    if 1<repov<=5:
        assess_total += 3
    elif 5<repov<=10:
        assess_total += 8
    elif 10<repov<=15:
        assess_total += 12
    elif 15<repov<=20:
        assess_total += 19
    elif 20<repov<=30:
        assess_total += 25
    elif repov>30:
        assess_total += 30

    #参与代码库评分
    if codcommitv>1 and codcommitv<=10000:
        assess_total += 3
    elif codcommitv>10001 and codcommitv<=30000:
        assess_total += 8
    elif codcommitv>30001 and codcommitv<=40000:
        assess_total += 13
    elif codcommitv>40001 and codcommitv<=50000:
        assess_total += 18
    elif codcommitv>50001 and codcommitv<=60000:
        assess_total += 23
    elif codcommitv>60001 and codcommitv<=70000:
        assess_total += 28
    elif codcommitv>70001 and codcommitv<=80000:
        assess_total += 32
    elif codcommitv>80001 and codcommitv<=100000:
        assess_total += 35
    elif codcommitv>100001 and codcommitv<=150000:
        assess_total += 37
    elif codcommitv>150001 and codcommitv<=200000:
        assess_total += 38
    elif codcommitv>2000001 and codcommitv<=500000:
        assess_total += 39
    elif codcommitv>500000:
        assess_total += 40

    return assess_total
    
        
```

```python

temp = []
for name,value in frame.groupby(frame['姓名']):
    newvalue = value.drop_duplicates(subset='代码库')

    info = newvalue.iloc[0]
    prosum = newvalue['项目ID'].unique()
    sum_attend = len(prosum)
    sum_repo = len(newvalue['代码库'])
    sum_code_commit = sum(newvalue['提交行数'])
    
    temp.append([name,info.iloc[1],info.iloc[2],assessment(sum_attend,sum_repo,sum_code_commit),sum_attend,sum_repo,sum_code_commit])
df = pd.DataFrame(temp,columns=['姓名','部门','岗位','总分','项目数','代码库数','代码提交行数'])  

print(df)
df.to_csv('./test.csv')
```
